```python
import time
from iceaggr.data.dataset import get_dataloader
```

```python
n_batches = 100
max_events = 50_000_000
cache_size = 1
```

```python
dataloader = get_dataloader(
    split="train",
    batch_size=2048,
    num_workers=0,
    max_events=max_events,
    collate_fn="deepsets",
    use_batch_aware_sampler=True,
    cache_size=cache_size,
)
```

```python
tick = time.time()
for i, batch in enumerate(dataloader):
    if i >= n_batches:
        break
tock = time.time()
print(f"Time taken: {tock - tick:.2f} seconds")
```
